Creation And DataTypes

```python
import torch
#creating tensor

a=torch.empty(2,3) #memory mai space dedega itne ka , jo already memory value mai hoga woh print kardega
print(a)
print(type(a))

#zeroes tensor
z=torch.zeros(4,5)
print(z)

#ones tensors
o=torch.ones(1,3)
print(o)

#random tensor
r=torch.rand(2,4)
print(r)
#use seed and then will always give same rand values
torch.manual_seed(100) # always do with imports so consistent in whole script

#othermethods, we have to use print as they return python object
print(torch.arange(0,10,2))
print(torch.linspace(0,5,5))
print(torch.eye(2))
print(torch.full((3,3),4))

#manually we can create
x=torch.tensor([[1,2,3],[4,5,6]])
print(x)
print(x.shape)

#can create of same size as any (zeroes,one etc)(rand will only work with float)
copy=torch.zeros_like(x)


#TensorDataTypes
print(x.dtype)

#ToCreate Integer Tensor some eg
Qa=torch.tensor([1.0,2.0,4.0],dtype=torch.int32)
Qb=torch.tensor([1.0,2.0,4.0],dtype=torch.int64)
Qc=torch.tensor([1.0,2.0,4.0],dtype=torch.float64)


#for already constrycted tensor,
Qd=Qc.to(torch.float64)
print(Qa,Qb,Qc,Qd)
```

Operations

```python
# scalar operations
s=torch.tensor([[1,2,3],[4,5,6]])
print(s+2)
print(s-2)
print(s*2)
print(s/2)
print(s**2) 
print(s%2)
print(s*100/2+25)
```

```python
#Elementwise operations
import torch

a=torch.tensor([[1,2,3],[4,5,6]])
b=torch.tensor([[7,8,9],[10,11,12]])

#some of eg
print(a+b)
print(a-b)
print(a%2)

c=torch.tensor([[1,2],[4,5]])
#print(a+c) #error or forever loading because size mismatch

#general operations
D=torch.tensor([[-1,2,-3],[4,-5,6]])
print(torch.abs(D)) #absolute value
print(torch.sqrt(torch.tensor([4.0,9.0,16.0]))) #sqrt only for float
print(torch.pow(a,2)) #square of each element
print(torch.sum(a)) #sum of all elements
print(torch.sum(a,dim=0)) #sum of each column
print(torch.sum(a,dim=1)) #sum of each row
print(torch.mean(torch.tensor([1.0,2.0,3.0,4.0]))) #mean only for float
print(torch.max(a)) #max element
print(torch.min(a)) #min element
print(torch.argmax(a)) #index of max element 
print(torch.argmin(a)) #index of min element
print(torch.transpose(a,0,1)) #transpose of matrix
print(a.T) #another way to transpose
print(torch.flatten(a)) #convert matrix to 1D tensor


#copying tensors
x=torch.tensor([[1,2,3],[4,5,6]])
y=x  # y is reference to x
y[0,0]=100
print(x) #x also changes because y is reference to x

#without reference
X=torch.tensor([[1,2,3],[4,5,6]])
Y=X.clone() #Y is copy of X
Y[0,0]=100
print(X) #X remains unchanged
print(Y)
```

```python
#Reshaping tensors
x=torch.tensor([[1,2,3],[4,5,6]])
print(x.shape)
y=x.view(3,2) #reshaped to 3 rows and 2 columns
print(y)
print(y.shape)

# Extracting elements
x=torch.tensor([[10,20,30],[40,50,60]])
print(x[0,1]) #20
print(x[1,:]) #entire 2nd row
print(x[:,2]) #entire 3rd column
```
